```python
import pandas as pd
```

```python
df = pd.read_csv('final_energy_set.csv')
df
```

```python
# Data for metal_cat_dict was obtained using the Nomad DataBase
# The calculations were carried out by Guo-xu Zhang
metal_cat_dict = {
    'Ca': [-18512.20035, -18512.20070, -18512.05603, -18512.20035],
    'Mg': [-5449.81766, -5449.81759, -5449.81759, -5449.77005],
    'Ba': [-221366.69618, -221366.69618, -221366.39988, -221366.69637]
}
metcat_df = pd.DataFrame(metal_cat_dict)
metcat_df
```

```python
metal_df_mean = pd.DataFrame({
    'Ca': [metcat_df['Ca'].mean()],
    'Mg': [metcat_df['Mg'].mean()],
    'Ba': [metcat_df['Ba'].mean()]
})
metal_df_mean
```

```python
df['Ca interaction energy'] = df['Ca interaction energy'] - metal_df_mean['Ca'].iloc[0]
df['Mg interaction energy'] = df['Mg interaction energy'] - metal_df_mean['Mg'].iloc[0]
df['Ba interaction energy'] = df['Ba interaction energy'] - metal_df_mean['Ba'].iloc[0]
df
```

```python
df.to_csv('energy_data.csv')
```
