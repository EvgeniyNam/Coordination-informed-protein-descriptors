```python
import pandas as pd
```

```python
df = pd.read_csv('Total_energy_raw.csv')
```

```python
df['Aminoacid'] = df['Mainfile'].apply(lambda x: x.split('/')[1])
df['Cation'] = df['Mainfile'].apply(lambda x: x.split('/')[3])

# Отображаем новый датасет
new_df = df[['Aminoacid', 'Cation', 'Total_energy']]
print(new_df)
```

```python
energy_df = new_df.groupby(['Aminoacid', 'Cation'], as_index=False)['Total_energy'].mean()
energy_df.head(10)
```

```python
columns = ['Aminoacid', 'Ca interaction energy', 'Mg interaction energy', 'Ba interaction energy']
energy_descriptor = pd.DataFrame(index = range(26), columns = columns)

n = 0

for i in range(26):
  energy_descriptor['Aminoacid'][i] = energy_df['Aminoacid'][n]
  energy_descriptor['Ba interaction energy'][i] = energy_df['Total_energy'][n] - energy_df['Total_energy'][n+3]
  energy_descriptor['Ca interaction energy'][i] = energy_df['Total_energy'][n+1] - energy_df['Total_energy'][n+3]
  energy_descriptor['Mg interaction energy'][i] = energy_df['Total_energy'][n+2] - energy_df['Total_energy'][n+3]
  n += 4

energy_descriptor['Aminoacid'][11] = 'his'
energy_descriptor = energy_descriptor[energy_descriptor['Aminoacid'].apply(lambda x: len(x) <= 3)].reset_index()

energy_descriptor.drop('index', axis=1, inplace=True)

energy_descriptor
```

```python
dict_aminoacids = {'ala': 'A', 'arg': 'R', 'asn': 'N', 'asp': 'D', 'val': 'V', 'his': 'H', 'gly': 'G', 'gln': 'Q', 'glu': 'E', 'ile': 'I',
                   'leu': 'L', 'lys': 'K', 'met': 'M', 'pro': 'P', 'ser': 'S', 'tyr': 'Y', 'thr': 'T', 'trp': 'W', 'phe': 'F', 'cys': 'C'}
```

```python
energy_descriptor['Aminoacid'] = energy_descriptor['Aminoacid'].replace(dict_aminoacids)
energy_descriptor
```

```python
energy_descriptor.to_csv('energy_set.csv', index = False)
```
